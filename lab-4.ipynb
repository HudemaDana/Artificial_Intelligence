```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + torch.exp(-x))

x1 = torch.tensor([1.0], requires_grad=True)
x2 = torch.tensor([-1.0], requires_grad=True)
y1 = sigmoid(x1)
y2 = sigmoid(x2)

y1.backward()
y2.backward()

print("The gradient of the sigmoid funciton is: ", x1.grad)
print("The gradient of the sigmoid funciton is: ", x2.grad)

ydx1 = sigmoid(x1) * (1 - sigmoid(x1))
ydx2 = sigmoid(x2) * (1 - sigmoid(x2))

print("The gradient of the sigmoid funciton is: ", ydx1)
print("The gradient of the sigmoid funciton is: ", ydx2)


# plt.plot(values, sigmoid(values))
# plt.xlabel('x')
# plt.ylabel('sigmoid(X)')
# plt.show()
```
